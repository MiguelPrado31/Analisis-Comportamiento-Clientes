### **Analisis Exploratorio de Datos con Python**

```python
# importar la libreria Pandas 
import pandas as pd
```

```python
# Cargar el dataset a la libreta jupyter
df = pd.read_csv("./analisis_compras_clientes.csv")
df.iloc[::100].head(10) # Se utilizo iloc[::100] con el fin de poder ver diferentes partes del Dataset y no solo el inicio
```

```python
# Usar df.info para revisar por encima la estructura del dataset 
df.info()
```

```python
# Resumen estadistico de todas las columnas usando .describe() 
df.describe(include='all')
```

```python
# Revisar las columnas donde hay 'null' en el dataset 
df.isnull().sum()
```

```python
""" 
Agrupar las Reviews por Categoria:
- De esta manera se podra calcular facilmente la Mediana de las Reviews de acuerdo a su Categoria
- Una vez con la Mediana de las Categorias, remplazarlos por los valores que faltan a traves de 
la funcion lambda. 
"""

df['Calificacion de Reviews'] = df.groupby('Categoria')['Calificacion de Reviews'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df.isnull().sum()
```

```python
# Hacer todo minuscula y en vez de espacios utilizar guiones bajo 
# Tambien conocido como: Snake Case
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

```python
# Quitar (usd) en Cantidad de Compra y errores de espacio de las columnas
df = df.rename(columns={'cantidad_de_compra_(usd)':'cantidad_de_compra'})
df = df.rename(columns={'descuento_aplicado_':'descuento_aplicado'})
df.columns
```

```python
# Crear nueva columna 'edad_grupo' con los siguientes labels
labels = ['Joven', 'Adulto', 'Mediana Edad', 'Anciano(a)']
# Segmentar la columna en 4 partes iguales y asignar los labels a traves de pd.qcut 
df['edad_grupo'] = pd.qcut(df['edad'], q=4, labels=labels)
```

```python
# Revisar la agrupacion de acuerdo a los labels
df[['edad_grupo', 'edad']].iloc[::20].head(10)
```

```python
# Nueva Columna para asignar los dias de la frecuencia de las compras
# a traves de la funcion .map de pandas
mapeo_frecuencia = {
    'Quincena': 15,  
    'Semanalmente': 7,
    'Mensualmente': 30,
    'Trimestral': 90,
    'Cada 2 semanas': 14,
    'Anualmente': 365,
    'Cada 3 Meses': 90

}


df['dias_frecuencias_compras'] = df['frecuencia_de_compras'].map(mapeo_frecuencia)
```

```python
# Revisar que se hayan asiganado a sus respectivos grupos
df[['frecuencia_de_compras', 'dias_frecuencias_compras']].head(10)
```

```python
df['dias_frecuencias_compras'].unique()
```

```python
df['dias_frecuencias_compras'].dtype
```

```python
df.columns
```

```python
# Necesitamos ambas columnas? Ambas columnas tienen los mismos valores?
df[['descuento_aplicado','promocion_usada']].iloc[::700].head(5) 
```

```python
# Revisar si hay diferencias entre las columnas
(df['descuento_aplicado'] == df['promocion_usada']).all()
```

```python
# Ya que es verdadero se removera la columna de promocion_usada
df = df.drop('promocion_usada', axis=1) # Remover la columna (1 = columna, 0 = fila)
df.columns # Revisar que se quito 
```

```python
# Revisar el dataset modificado, cada 10 valores
df.iloc[::10]
```

```python
# Guardar el dataset modificado en un XLSX 
df.to_excel('analisis_clientes_modificado.xlsx', index=False)
```
